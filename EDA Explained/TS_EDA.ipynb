# Taylor Swift - EDA

```python
# Import necessary libraries
import numpy as np  # NumPy for numerical operations
import pandas as pd  # Pandas for handling tabular data
import seaborn as sns  # Seaborn for data visualization
import matplotlib.pyplot as plt  # Matplotlib for plotting
from sklearn.preprocessing import OrdinalEncoder  # OrdinalEncoder for encoding categorical variables

# Load the dataset into a Pandas DataFrame
TS = pd.read_csv("taylor_swift_spotify.csv")  # Reads the CSV file into a DataFrame

# Display the first few rows of the dataset (in Jupyter Notebook, 'display' is preferred over 'print')
display(TS)

# Print a summary of the dataset, including column names, non-null values, and data types
TS.info()
```

## Heatmap

```python
# Set the figure size for the plot
plt.figure(figsize=(12, 6))  # Creates a figure with a width of 12 inches and height of 6 inches

# Set the Seaborn style for better aesthetics
sns.set(style="whitegrid")  # Uses a white grid style to make the visualization clearer

# Compute the correlation matrix, dropping unnecessary columns
corr = TS.drop(columns=["Unnamed: 0", "track_number", "name", "album", "release_date", "id", "uri"]).corr()  
# Drops the "Unnamed: 0" and "track_number" columns as they may not be useful for correlation analysis
# Computes the correlation matrix for the remaining numerical features

# Create a heatmap to visualize the correlation matrix
sns.heatmap(corr, annot=True, cmap="coolwarm")  
# `annot=True` adds numerical values inside the heatmap cells
# `cmap="coolwarm"` sets the color scheme from cool (blue) to warm (red)

# Rotate x-axis labels for better readability
plt.xticks(rotation=30)  # Rotates the x-axis labels by 30 degrees

# Display the heatmap
plt.show()  # Renders the heatmap visualization
```

```python
# Count the number of occurrences of each unique album in the dataset
TS['album'].value_counts()  
# This returns a series where the index is the album name, and the values are the count of songs per album
# Useful for understanding the distribution of songs across different albums
```

## Boxplot

```python
# Collect all features having [0, 1] as the range of values
melted_TS = pd.melt(TS, id_vars="album", value_vars=TS.columns.values[np.r_[7:12, 13, 15]])
# `pd.melt()` reshapes the DataFrame from wide format to long format for better visualization
# `id_vars="album"` keeps the "album" column as an identifier
# `value_vars=TS.columns.values[np.r_[7:12, 13, 15]]` selects specific feature columns 
# that have values in the range [0,1]

# Rename columns for clarity
melted_TS.columns = ['Album', 'Features', 'Values']

# Display the reshaped dataset
display(melted_TS)

# Plot boxplots per album to compare feature distributions
plt.figure(figsize=(10, 20))  # Set figure size for better visualization

# Create a horizontal boxplot to show the distribution of features per album
sns.boxplot(data=melted_TS, width=0.8, orient="h",  
            x="Values", y="Features", hue='Album')  
# `x="Values"`: Represents the feature values on the x-axis
# `y="Features"`: Lists the features on the y-axis
# `hue='Album'`: Differentiates each album with different colors

# Adjust label size for better readability
plt.rc('axes', labelsize=15)

# Move the legend outside the plot for better visualization
plt.legend(bbox_to_anchor=(1, -0.05))

# Add grid lines for better readability
plt.grid()

# Show the plot
plt.show()
```

```python
# Set figure size for better visualization
plt.figure(figsize=(10, 15))  # Creates a figure with a width of 10 inches and height of 15 inches

# Convert song duration from milliseconds to minutes
TS["duration_min"] = TS["duration_ms"] / 60000  
# `duration_ms` is originally in milliseconds; dividing by 60000 converts it to minutes

# Create a boxplot to visualize song duration distribution per album
sns.boxplot(data=TS, x="duration_min", y="album")  
# `x="duration_min"`: Displays the song duration in minutes on the x-axis
# `y="album"`: Lists the albums on the y-axis to compare song durations across albums

# Add grid lines for better readability
plt.grid()

# Display the plot
plt.show()

# Remove the temporary 'duration_min' column after visualization
TS.drop(columns="duration_min", inplace=True)  
# Drops the column to keep the original dataset unchanged
```

```python
# Set the figure size for better visualization
plt.figure(figsize=(10, 15))  # Creates a figure with a width of 10 inches and height of 15 inches

# Create a boxplot to visualize the distribution of song popularity per album
sns.boxplot(data=TS, x="popularity", y="album")  
# `x="popularity"`: Shows song popularity scores on the x-axis
# `y="album"`: Lists the albums on the y-axis for comparison
# The boxplot displays the distribution of popularity scores for each album

# Add grid lines to improve readability
plt.grid()

# Show the plot
plt.show()
```

## Rankings

```python
# Iterate over all columns starting from index 7 to the last column
for prop in TS.columns.values[7:]:  
    print(f'Ranking by {prop}')  # Print the feature being ranked

    # Sort the dataset by the current feature in descending order
    sorted_TS = TS.sort_values(by=prop, ascending=False)[["name", prop]]  
    # Select only the "name" column (song title) and the current feature

    # Remove duplicate song names to ensure unique entries in the ranking
    display(sorted_TS.drop_duplicates(subset='name'))  
    # `drop_duplicates(subset='name')` ensures that each song appears only once
```

## Frequency of release dates

```python
# Convert the "release_date" column to a datetime object for easier manipulation
TS['release_date'] = pd.to_datetime(TS['release_date'])

# Extract the day, month, and year from the release date
month_of_release = TS['release_date'].dt.month  # Extracts the month (1-12)
day_of_month = TS['release_date'].dt.day  # Extracts the day of the month (1-31)
year_of_release = TS['release_date'].dt.year  # Extracts the year

# Create subplots for visualizing the distribution of release dates
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 4), sharey=False)
# `nrows=1, ncols=3`: Creates a row with 3 separate plots
# `figsize=(20,4)`: Makes the figure 20 inches wide and 4 inches tall
# `sharey=False`: Allows each subplot to have its own y-axis scale

# Plot the count of songs released on each day of the month
sns.countplot(x=day_of_month, ax=axes[0])  
# `x=day_of_month`: Groups songs by their day of release and counts occurrences
# `ax=axes[0]`: Assigns this plot to the first subplot

# Plot the count of songs released in each month
sns.countplot(x=month_of_release, ax=axes[1])  
# `x=month_of_release`: Groups songs by their month of release and counts occurrences
# `ax=axes[1]`: Assigns this plot to the second subplot

# Plot the count of songs released per year
sns.countplot(x=year_of_release, ax=axes[2])  
# `x=year_of_release`: Groups songs by release year and counts occurrences
# `ax=axes[2]`: Assigns this plot to the third subplot

# Set titles for each subplot
axes[0].set_title('Day of Month Released')  
axes[1].set_title('Month of Release')  
axes[2].set_title('Year of Release')  

# Show the final visualization
plt.show()
```

## Evolutions of Data with time

```python
# Add a new column for the release year extracted earlier
TS["release_year"] = year_of_release  

# Iterate through selected features (columns 7 to 15) to analyze their trend over the years
for prop in TS.columns.values[7:16]:  
    # Compute the mean of the current feature for each release year
    means = TS.groupby("release_year")[prop].mean()  
    # `groupby("release_year")`: Groups songs by their release year
    # `[prop].mean()`: Computes the average value of the feature for each year

    # Create a new figure for each feature's trend
    plt.figure(figsize=(8, 3))  # Set figure size for better visualization

    # Plot the trend of the feature over the years
    plt.plot(means.index, means.loc[:], 'bo-', linewidth=2)  
    # `means.index`: Represents the years on the x-axis
    # `means.loc[:]`: Represents the average values of the feature on the y-axis
    # `'bo-'`: Uses blue circles with a line connecting them
    # `linewidth=2`: Increases line thickness for better visibility

    # Set x-axis labels with years from 2006 to 2023, rotating them for clarity
    plt.xticks(np.arange(2006, 2024), rotation=30)  

    # Label the y-axis with the feature name
    plt.ylabel(f"Average {prop}")  

    # Label the x-axis as 'Year'
    plt.xlabel('Year')  

    # Show the plot
    plt.show()  
```
